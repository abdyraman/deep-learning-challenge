## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
cleaned_df = application_df.drop(['EIN', 'NAME'], axis=1)
cleaned_df
```

```python
# Determine the number of unique values in each column.
unique_counts = cleaned_df.nunique()
unique_counts
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_type_counts = cleaned_df['APPLICATION_TYPE'].value_counts()
app_type_counts
```

```python
cutoff_value = 500
application_types_to_replace = app_type_counts[app_type_counts < cutoff_value].index

# Replace in the dataframe
for app in application_types_to_replace:
    cleaned_df.loc[cleaned_df['APPLICATION_TYPE'] == app, 'APPLICATION_TYPE'] = "Other"

# Check the updated value counts
application_type_counts = cleaned_df['APPLICATION_TYPE'].value_counts()
print(application_type_counts)
```

```python
# You may find it helpful to look at CLASSIFICATION value
classification_counts = cleaned_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
filtered_counts = classification_counts[classification_counts > 1]
print(filtered_counts)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value = 1000
classifications_to_replace = classification_counts[classification_counts < cutoff_value].index

# Replace in dataframe
for cls in classifications_to_replace:
    cleaned_df['CLASSIFICATION'] = cleaned_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
cleaned_df['CLASSIFICATION'].value_counts()
```

```python
cleaned_df.dtypes
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
X = pd.get_dummies(cleaned_df)
X.shape
```

```python
X = X.drop(['IS_SUCCESSFUL'], axis=1)
```

```python
X
```

```python
# Split our preprocessed data into our features and target arrays
y=application_df["IS_SUCCESSFUL"].values.reshape(-1, 1)

# Split the preprocessed data into a training and testing dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train_scaled
# X_test_scaled
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
# First hidden layer
# Second hidden layer
# Output layer

nn = tf.keras.models.Sequential()
nn.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=43))
nn.add(tf.keras.layers.Dense(units=80, activation="relu"))
nn.add(tf.keras.layers.Dense(units=80, activation="tanh"))
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the Sequential model
nn.summary()
```

```python
# Train the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
fit_model = nn.fit(X_train_scaled, y_train, epochs=50)
```

```python
# Compile the model
model_loss, model_accuracy = nn.evaluate(X_train_scaled,y_train,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
